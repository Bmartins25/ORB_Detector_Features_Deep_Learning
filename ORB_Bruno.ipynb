```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("teste_surf.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
img = cv2.imread("teste_surf.jpeg", cv2.IMREAD_GRAYSCALE)
```

```python
orb = cv2.ORB_create()
```

```python
keypoints_ref, descriptors_ref = orb.detectAndCompute(img, None)
keypoints_target, descriptors_target = orb.detectAndCompute(img, None)
```

```python
matcher = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
```

```python
matches = matcher.match(descriptors_ref, descriptors_target)
```

```python
matches = sorted(matches, key=lambda x: x.distance)
```

```python
 img_matches = cv2.drawMatches(
        img, keypoints_ref,
        img, keypoints_target,
        matches[:10], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS
    )
```

```python
plt.imshow(img),plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
